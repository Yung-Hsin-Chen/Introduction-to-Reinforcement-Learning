```python
plt.figure(figsize = (20, 5))
N_v = 120000
N_q = 80000

## SARSA TABULAR

plot_lst_v1 = [R_v1[i]/N_v1[i] for i in range(N_v)]
plot_lst_v2 = [R_v2[i]/N_v2[i] for i in range(N_v)]
plot_lst_v3 = [R_v3[i]/N_v3[i] for i in range(N_v)]
plot_lst_v4 = [R_v4[i]/N_v4[i] for i in range(N_v)]
plot_lst_v5 = [R_v5[i]/N_v5[i] for i in range(N_v)]
plot_lst_v6 = [R_v6[i]/N_v6[i] for i in range(N_v)]
plot_lst_v9 = [R_v9[i]/N_v9[i] for i in range(N_v)]
plot_lst_v10 = [R_v10[i]/N_v10[i] for i in range(N_v)]

res_data_v = {"v1":plot_lst_v1, "v2":plot_lst_v2, "v3":plot_lst_v3, "v4":plot_lst_v4, "v5":plot_lst_v5, "v6":plot_lst_v6, "v9":plot_lst_v9, "v10":plot_lst_v10}

df_v = pd.DataFrame.from_dict(res_data_v)

plot_v1 = df_v["v1"].rolling(window = 1000, min_periods = 1).mean()
plot_v2 = df_v["v2"].rolling(window = 1000, min_periods = 1).mean()
plot_v3 = df_v["v3"].rolling(window = 1000, min_periods = 1).mean()
plot_v4 = df_v["v4"].rolling(window = 1000, min_periods = 1).mean()
plot_v5 = df_v["v5"].rolling(window = 1000, min_periods = 1).mean()
plot_v6 = df_v["v6"].rolling(window = 1000, min_periods = 1).mean()
plot_v9 = df_v["v9"].rolling(window = 1000, min_periods = 1).mean()
plot_v10 = df_v["v10"].rolling(window = 1000, min_periods = 1).mean()

## Q-LEARNING
plot_lst_q1 = [R_q1[i]/N_q1[i] for i in range(N_q)]
plot_lst_q3 = [R_q3[i]/N_q3[i] for i in range(N_q)]
plot_lst_q4 = [R_q4[i]/N_q4[i] for i in range(N_q)]
plot_lst_q7 = [R_q7[i]/N_q7[i] for i in range(N_q)]
plot_lst_q8 = [R_q8[i]/N_q8[i] for i in range(N_q)]

res_data_q = {"q1":plot_lst_q1, "q3":plot_lst_q3, "q4":plot_lst_q4, "q7":plot_lst_q7, "q8":plot_lst_q8}

df_q = pd.DataFrame.from_dict(res_data_q)

plot_q1 = df_q["q1"].rolling(window = 1000, min_periods = 1).mean()
plot_q3 = df_q["q3"].rolling(window = 1000, min_periods = 1).mean()
plot_q4 = df_q["q4"].rolling(window = 1000, min_periods = 1).mean()
plot_q7 = df_q["q7"].rolling(window = 1000, min_periods = 1).mean()
plot_q8 = df_q["q8"].rolling(window = 1000, min_periods = 1).mean()
```
